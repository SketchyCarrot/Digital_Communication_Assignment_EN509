```python
import matplotlib.pyplot as plt
import numpy as np
import sounddevice as sd
```

```python
DURATION = int(input("Duration of recording (in integers): "))
FS = 1000
```

```python
print("Recording...")
audio = sd.rec(int(DURATION * FS), samplerate=FS, channels=1, dtype='float64')
sd.wait()
print("Recording complete. Here is your audio recording: \n")

plt.figure(figsize=(10, 4))
plt.plot(np.linspace(0, DURATION, audio.shape[0]), audio)
plt.title("Audio Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
